```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
L1 = [(0.6634, 0.9046, 0.9055, 0.8445), (0.66, 0.90, 0.91, 0.85)]
classfier = ['Decision Tree', 'MultinomialNB', 'Linear SVC', 'XGBClassifier']
colors = ["#99CC01","#FFFF01","#0000FE","#FE0000","#A6A6A6","#D9E021"]
```

```python
x_val = [i for i in classfier]
y_val =  L1[0]

plt.bar(left=range(4),height=y_val,width=0.5,color=colors[0:4])
plt.xticks(range(4),x_val)
plt.title('Accuracy of Classifers')
plt.ylabel('Accuracy')
plt.show()
```

```python
x_val = [i for i in classfier]
y_val =  L1[1]

plt.bar(left=range(4),height=y_val,width=0.5,color=colors[0:4])
plt.xticks(range(4),x_val)
plt.title('F1-score of Classifers')
plt.ylabel('F1-score')
plt.show()
```
